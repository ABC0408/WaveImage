```python
from scipy.io.wavfile import write
import numpy as np
import cv2
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa.display
%matplotlib inline
```

```python
def np2pil(audio, d): # change the image size
    audio = Image.fromarray(audio)
    audio = audio.resize((d, d), resample=Image.ANTIALIAS)
    return np.array(audio).astype(np.float32)

def audio2pic(audio, path): # save audio to picture
    diff = len(audio) - np.square(int(np.sqrt(len(audio))))
    if diff %2 == 0:
        l, r = diff //2, diff //2
    else:
        l, r = diff //2, diff //2 + 1
    if diff:
        audio0 = np.reshape(audio[l:-r], (-1, int(np.sqrt(len(audio)))))
    else:
        audio0 = np.reshape(audio, (-1, int(np.sqrt(len(audio)))))
    audio0 = np2pil((audio0+1)/2, 512).T*255
    # scipy.misc.toimage(audio0, cmin=0, cmax=1).save('./images/speech-mel-14900-640.png')
    # audio1 = mpimg.imread('./images/speech-mel-14900-640.png')*255
    # audio1 = scipy.misc.imresize(audio1, (int(np.sqrt(len(audio))), int(np.sqrt(len(audio)))), interp='bicubic')#.reshape(-1).T/255
    audio2 = np.zeros((512, 512, 3))
    audio2[:, :, 0] = audio0
    audio2[:, :, 1] = audio0
    audio2[:, :, 2] = audio0
    cv2.imwrite(path, audio2, [int(cv2.IMWRITE_PNG_COMPRESSION), 0])
    return audio2
```

```python
audio = np.load('speech-audio-14901.npy') # load original speech file
audio.shape
```

```python
ipd.Audio(audio, rate=48000)
```

```python
wav_image = audio2pic(audio, 'wave-image.png')
wav_image.shape
```

```python
plt.imshow(wav_image[:, :, 0], cmap='gray')
```

```python
wav_image = np2pil(wav_image[:, :, 0], 328)/255 # pixel range convert to [0, 1]
wav_image = wav_image.T.reshape(-1)* 2 - 1
wav_image.shape
```

```python
ipd.Audio(wav_image, rate=48000)
```

```python
librosa.display.waveplot(wav_image, sr=48000)
```
